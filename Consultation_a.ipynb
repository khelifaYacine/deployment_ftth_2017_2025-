## Étapes pour corriger les types de variables
### Correction et reconvertion des types de variables 



```python
# Conversion des dates
df["debut_deploiement"] = pd.to_datetime(df["debut_deploiement"], errors="coerce")
df["fin_deploiement_amii"] = pd.to_datetime(df["fin_deploiement_amii"], errors="coerce")
df["fin_déploiement_l33-13"] = pd.to_datetime(df["fin_déploiement_l33-13"], errors="coerce")
df["date_timer_ref"] = pd.to_datetime(df["date_timer_ref"], errors="coerce")

# Conversion des périodes en string ou catégorie
df["semestre"] = df["semestre"].astype("category")
df["trimestre"] = df["trimestre"].astype("category")
df["mois"] = df["mois"].astype("Int64")  # pour garder les NaN
df["annee"] = df["annee"].astype("Int64")

# Conversion des identifiants et géos si besoin
df["code_insee"] = df["code_insee"].astype("str")
df["geom"] = df["geom"].astype("string")
df["centroid"] = df["centroid"].astype("string")
```

## Data Visualisation 

```python
# ANalyse de la somme des locaux raccordables par communes

df_commune = df.groupby("commune")["locaux_raccordables"].sum().reset_index()
df_commune = df_commune.sort_values(by="locaux_raccordables", ascending=False)
df_commune
```

```python
# Sélection des 5 premières et 5 dernières
top5 = df_commune.head(5)
bottom5 = df_commune.tail(5)

# Affichage
print("🔝 Top 5 des communes les plus raccordées :")
print(top5)

print("\n🔻 Bottom 5 des communes les moins raccordées :")
print(bottom5)
```

```python

# Tri pour affichage clair
df_plot = pd.concat([top5, bottom5])
df_plot = df_plot.sort_values(by="locaux_raccordables", ascending=True)

# Couleurs
colors = ['red'] * 5 + ['green'] * 5

# Graphique
plt.figure(figsize=(10,6))
bars = plt.barh(df_plot["commune"], df_plot["locaux_raccordables"], color=colors)
plt.xscale('log')  # Échelle logarithmique
plt.xlabel("Nombre de locaux raccordables (échelle log)")
plt.title("Top 5 vs Bottom 5 des communes raccordées en FTTH")

# Affichage des valeurs
for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.05, bar.get_y() + bar.get_height()/2,
             f'{int(width):,}', va='center')

plt.tight_layout()
plt.show()
```

### Identifier la variable de raccordement réel


```python
df["locaux_raccordés"] = df["locaux_meilleure_estimation"] * df["part_raccordables_estim"]

# Regroupement par année
df_evolution = df.groupby("annee")["locaux_raccordés"].sum().reset_index()

# Visualisation

plt.figure(figsize=(10,6))
plt.plot(df_evolution["annee"], df_evolution["locaux_raccordés"], marker='o', color='blue')
plt.title("Évolution annuelle des locaux raccordés FTTH")
plt.xlabel("Année")
plt.ylabel("Nombre de locaux raccordés")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# op 5 / Bottom 5 des communes raccordées
df_commune_raccordés = df.groupby("commune")["locaux_raccordés"].sum().reset_index()
df_commune_raccordés = df_commune_raccordés.sort_values(by="locaux_raccordés", ascending=False)

top5 = df_commune_raccordés.head(5)
bottom5 = df_commune_raccordés.tail(5)
df_plot = pd.concat([top5, bottom5])
df_plot = df_plot.sort_values(by="locaux_raccordés", ascending=True)

# Étape 5 : Couleurs
colors = ['red'] * 5 + ['green'] * 5  # rouge pour les moins bons, vert pour les meilleurs

# Étape 6 : Graphique
plt.figure(figsize=(10,6))
bars = plt.barh(df_plot["commune"], df_plot["locaux_raccordés"], color=colors)
plt.xlabel("Nombre de locaux raccordés")
plt.title("Top 5 vs Bottom 5 des communes raccordées en FTTH")

# Ajout des valeurs sur les barres
for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2,
             f'{int(width):,}', va='center')

plt.tight_layout()
plt.show()
```

```python
print("📌 KPI FTTH déploiement :")
print(f"Total raccordés : {int(df['locaux_raccordés'].sum()):,}")
print(f"Taux moyen de raccordement : {df['part_raccordables_estim'].mean():.2%}")
print(f"Nombre de communes : {df['commune'].nunique()}")
print(f"Année la plus active : {df_evolution.sort_values(by='locaux_raccordés', ascending=False).iloc[0]['annee']}")
```

#### Calculer les locaux non raccordés

```python
# Étape 1 : Nettoyage du ratio
df["part_raccordables_estim"] = df["part_raccordables_estim"].clip(lower=0, upper=1)

# Étape 2 : Calcul des locaux raccordés
df["locaux_raccordés"] = df["locaux_meilleure_estimation"] * df["part_raccordables_estim"]

# Étape 3 : Agrégation par commune
df_commune = df.groupby("commune")[["locaux_meilleure_estimation", "locaux_raccordés"]].sum().reset_index()

# Étape 4 : Calcul des locaux restants
df_commune["locaux_non_raccordés"] = df_commune["locaux_meilleure_estimation"] - df_commune["locaux_raccordés"]

# Étape 5 : Filtrer les communes avec des locaux restants
df_restants = df_commune[df_commune["locaux_non_raccordés"] > 0]

# Étape 6 : Affichage
print("📌 Communes avec des locaux restant à raccorder :")
print(df_restants[["commune", "locaux_non_raccordés"]].sort_values(by="locaux_non_raccordés", ascending=False))
```

```python
# Top 10 des communes les plus en retard
top10_restants = df_restants.sort_values(by="locaux_non_raccordés", ascending=False).head(10)
top10_restants.head(15)
```

```python
# 2. KPI global
total_restants = df_restants["locaux_non_raccordés"].sum()
print(f"🔢 Total de locaux restant à raccorder : {int(total_restants):,}")
```

```python
top10 = df_restants.sort_values(by="locaux_non_raccordés", ascending=False).head(10)

plt.figure(figsize=(10,6))
bars = plt.barh(top10["commune"], top10["locaux_non_raccordés"], color="orange")
plt.xlabel("Locaux restant à raccorder")
plt.title("Top 10 des communes en retard de raccordement")

for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2,
             f'{int(width):,}', va='center')

plt.tight_layout()
plt.show()
```

### comparer les communes en retard avec leur fin_déploiement_l33-13 pour détecter les retards par rapport au planning

```python
df["fin_déploiement_l33-13"] = pd.to_datetime(df["fin_déploiement_l33-13"], errors="coerce")
df_retard = df[df["locaux_non_raccordés"] > 0]
df_retard = df_retard[df_retard["fin_déploiement_l33-13"] < pd.Timestamp("today")]
df_retard.head() 
```

```python
total_raccordables = df["locaux_meilleure_estimation"].sum()
total_raccordés = df["locaux_raccordés"].sum()
total_restants = total_raccordables - total_raccordés
taux_couverture = total_raccordés / total_raccordables

print(f"🔢 Total raccordables : {int(total_raccordables):,}")
print(f"✅ Total raccordés : {int(total_raccordés):,}")
print(f"⚠️ Total restant à raccorder : {int(total_restants):,}")
print(f"📈 Taux de couverture FTTH : {taux_couverture:.2%}")
```

```python
import matplotlib.pyplot as plt

# Données
indicateurs = ['Total Raccordables', 'Total Raccordés', 'Total Restant à Raccorder', 'Taux de Couverture (%)']
valeurs = [total_raccordables, total_raccordés, total_restants, taux_couverture]
couleurs = ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728']

# Affichage
plt.figure(figsize=(8,6))
bars = plt.bar(indicateurs, valeurs, color=couleurs)

# Ajout des valeurs sur les barres
for i, bar in enumerate(bars):
    height = bar.get_height()
    label = f'{int(height):,}' if i < 3 else f'{height:.2f}%'
    plt.text(bar.get_x() + bar.get_width()/2, height * 1.01, label, ha='center', va='bottom')

plt.title("Indicateurs Globaux du FTTH")
plt.ylabel("Valeurs")
plt.tight_layout()
plt.show()
```
