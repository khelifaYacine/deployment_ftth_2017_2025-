## √âtapes pour corriger les types de variables
### Correction et reconvertion des types de variables 



```python
# Conversion des dates
df["debut_deploiement"] = pd.to_datetime(df["debut_deploiement"], errors="coerce")
df["fin_deploiement_amii"] = pd.to_datetime(df["fin_deploiement_amii"], errors="coerce")
df["fin_d√©ploiement_l33-13"] = pd.to_datetime(df["fin_d√©ploiement_l33-13"], errors="coerce")
df["date_timer_ref"] = pd.to_datetime(df["date_timer_ref"], errors="coerce")

# Conversion des p√©riodes en string ou cat√©gorie
df["semestre"] = df["semestre"].astype("category")
df["trimestre"] = df["trimestre"].astype("category")
df["mois"] = df["mois"].astype("Int64")  # pour garder les NaN
df["annee"] = df["annee"].astype("Int64")

# Conversion des identifiants et g√©os si besoin
df["code_insee"] = df["code_insee"].astype("str")
df["geom"] = df["geom"].astype("string")
df["centroid"] = df["centroid"].astype("string")
```

## Data Visualisation 

```python
# ANalyse de la somme des locaux raccordables par communes

df_commune = df.groupby("commune")["locaux_raccordables"].sum().reset_index()
df_commune = df_commune.sort_values(by="locaux_raccordables", ascending=False)
df_commune
```

```python
df["locaux_raccord√©s"] = df["locaux_meilleure_estimation"] * df["part_raccordables_estim"]

# Regroupement par ann√©e
df_evolution = df.groupby("annee")["locaux_raccord√©s"].sum().reset_index()

# Visualisation

plt.figure(figsize=(10,6))
plt.plot(df_evolution["annee"], df_evolution["locaux_raccord√©s"], marker='o', color='blue')
plt.title("√âvolution annuelle des locaux raccord√©s FTTH")
plt.xlabel("Ann√©e")
plt.ylabel("Nombre de locaux raccord√©s")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# op 5 / Bottom 5 des communes raccord√©es
df_commune_raccord√©s = df.groupby("commune")["locaux_raccord√©s"].sum().reset_index()
df_commune_raccord√©s = df_commune_raccord√©s.sort_values(by="locaux_raccord√©s", ascending=False)

top5 = df_commune_raccord√©s.head(5)
bottom5 = df_commune_raccord√©s.tail(5)
df_plot = pd.concat([top5, bottom5])
df_plot = df_plot.sort_values(by="locaux_raccord√©s", ascending=True)

# √âtape 5 : Couleurs
colors = ['red'] * 5 + ['green'] * 5  # rouge pour les moins bons, vert pour les meilleurs

# √âtape 6 : Graphique
plt.figure(figsize=(10,6))
bars = plt.barh(df_plot["commune"], df_plot["locaux_raccord√©s"], color=colors)
plt.xlabel("Nombre de locaux raccord√©s")
plt.title("Top 5 vs Bottom 5 des communes raccord√©es en FTTH")

# Ajout des valeurs sur les barres
for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2,
             f'{int(width):,}', va='center')

plt.tight_layout()
plt.show()
```

```python
print("üìå KPI FTTH d√©ploiement :")
print(f"Total raccord√©s : {int(df['locaux_raccord√©s'].sum()):,}")
print(f"Taux moyen de raccordement : {df['part_raccordables_estim'].mean():.2%}")
print(f"Nombre de communes : {df['commune'].nunique()}")
print(f"Ann√©e la plus active : {df_evolution.sort_values(by='locaux_raccord√©s', ascending=False).iloc[0]['annee']}")
```

#### Calculer les locaux non raccord√©s

```python
# √âtape 1 : Nettoyage du ratio
df["part_raccordables_estim"] = df["part_raccordables_estim"].clip(lower=0, upper=1)

# √âtape 2 : Calcul des locaux raccord√©s
df["locaux_raccord√©s"] = df["locaux_meilleure_estimation"] * df["part_raccordables_estim"]

# √âtape 3 : Agr√©gation par commune
df_commune = df.groupby("commune")[["locaux_meilleure_estimation", "locaux_raccord√©s"]].sum().reset_index()

# √âtape 4 : Calcul des locaux restants
df_commune["locaux_non_raccord√©s"] = df_commune["locaux_meilleure_estimation"] - df_commune["locaux_raccord√©s"]

# √âtape 5 : Filtrer les communes avec des locaux restants
df_restants = df_commune[df_commune["locaux_non_raccord√©s"] > 0]

# √âtape 6 : Affichage
print("üìå Communes avec des locaux restant √† raccorder :")
print(df_restants[["commune", "locaux_non_raccord√©s"]].sort_values(by="locaux_non_raccord√©s", ascending=False))
```

```python
# Top 10 des communes les plus en retard
top10_restants = df_restants.sort_values(by="locaux_non_raccord√©s", ascending=False).head(10)
top10_restants.head(15)
```

```python
# 2. KPI global
total_restants = df_restants["locaux_non_raccord√©s"].sum()
print(f"üî¢ Total de locaux restant √† raccorder : {int(total_restants):,}")
```

```python
top10 = df_restants.sort_values(by="locaux_non_raccord√©s", ascending=False).head(10)

plt.figure(figsize=(10,6))
bars = plt.barh(top10["commune"], top10["locaux_non_raccord√©s"], color="orange")
plt.xlabel("Locaux restant √† raccorder")
plt.title("Top 10 des communes en retard de raccordement")

for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2,
             f'{int(width):,}', va='center')

plt.tight_layout()
plt.show()
```

### comparer les communes en retard avec leur fin_d√©ploiement_l33-13 pour d√©tecter les retards par rapport au planning

```python
df["fin_d√©ploiement_l33-13"] = pd.to_datetime(df["fin_d√©ploiement_l33-13"], errors="coerce")
df_retard = df[df["locaux_non_raccord√©s"] > 0]
df_retard = df_retard[df_retard["fin_d√©ploiement_l33-13"] < pd.Timestamp("today")]
df_retard.head() 
```

```python
total_raccordables = df["locaux_meilleure_estimation"].sum()
total_raccord√©s = df["locaux_raccord√©s"].sum()
total_restants = total_raccordables - total_raccord√©s
taux_couverture = total_raccord√©s / total_raccordables

print(f"üî¢ Total raccordables : {int(total_raccordables):,}")
print(f"‚úÖ Total raccord√©s : {int(total_raccord√©s):,}")
print(f"‚ö†Ô∏è Total restant √† raccorder : {int(total_restants):,}")
print(f"üìà Taux de couverture FTTH : {taux_couverture:.2%}")
```

```python
import matplotlib.pyplot as plt

# Donn√©es
indicateurs = ['Total Raccordables', 'Total Raccord√©s', 'Total Restant √† Raccorder', 'Taux de Couverture (%)']
valeurs = [total_raccordables, total_raccord√©s, total_restants, taux_couverture]
couleurs = ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728']

# Affichage
plt.figure(figsize=(8,6))
bars = plt.bar(indicateurs, valeurs, color=couleurs)

# Ajout des valeurs sur les barres
for i, bar in enumerate(bars):
    height = bar.get_height()
    label = f'{int(height):,}' if i < 3 else f'{height:.2f}%'
    plt.text(bar.get_x() + bar.get_width()/2, height * 1.01, label, ha='center', va='bottom')

plt.title("Indicateurs Globaux du FTTH")
plt.ylabel("Valeurs")
plt.tight_layout()
plt.show()
```
